```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from colour import Color
```

```python
# graph 1 data
categories = ['2021', '2022', '2023']
agree = [18, 15, 10]
neutral = [45, 46, 36]
disagree = [37, 38, 52]

# graph 2 data
use_cases = ['To write essays', 'To solve math\nproblems', 'To research\nnew topics']
not_acceptable = [57, 36, 13]
acceptable = [20, 39, 69]
not_sure = [23, 24, 18]

# graph 3 data
source_healthcare = ['They would _____ if their\nhealth care provider\n relied on AI.', 'It would make them\n_____ with the health outcomes\nfor patients']
good = [39, 38]
bad = [60, 33]
eh = [0, 27]
no_answer = [1, 2]
empty = [0, 0]

# graph 4 data
source_labels_content_ai= ['DALL-E 2\n916 million', 'Midjourney\n964 million', 'Adobe Firefly\n1 billion', 'Models based on\nStable Diffusion\n12.590 billion']
sizes_content_ai = [9.16, 9.64, 10, 125.9]

# graph 5 data
source_labels_content_total= ['Shutterstock Library\n386 millions', 'AI-Created Images\n15 billion', 'Images Uploaded\n to Instagram\n50 billion', 'Images Indexed\n on Google\n136 billion', 'Pins Saved\n on Pinterest\n240 billion', 'Images Uploaded\n to Facebook\n1.5 trillion']
sizes_content_total= [3.86, 150, 500, 1360, 2400, 15000]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(categories))
ax.set_xticks(x)
ax.set_xticklabels(categories)

bar_width = 0.25
ax.bar(x - bar_width, disagree, width=bar_width, label='More Concerned', color='#FFA500', edgecolor='black')
ax.bar(x, agree, width=bar_width, label='Less Concerned', color='#0000FF', edgecolor='black')
ax.bar(x + bar_width, neutral, width=bar_width, label='Neutral', color='#808080', edgecolor='black')

ax.set_ylabel('Percentage')
ax.set_title('"Are you more or less concerned about AI use in daily life?"')
plt.yticks(np.arange(0, 101, 5))
ax.set_ylim(0, 55)
ax.legend()

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(use_cases))
ax.set_xticks(x)
ax.set_xticklabels(use_cases)

bar_width = 0.25
ax.bar(x - bar_width, not_acceptable, width=bar_width, label='Not acceptable', color='#FFA500', edgecolor='black')
ax.bar(x, acceptable, width=bar_width, label='Acceptable', color='#0000FF', edgecolor='black')
ax.bar(x + bar_width, not_sure, width=bar_width, label='Not sure', color='#808080', edgecolor='black')

ax.set_ylabel('Percentage (%)')
ax.set_title('U.S. Teens\' Perceptions of Acceptable ChatGPT Usage')
plt.yticks(np.arange(0, 101, 5))
ax.set_ylim(0, 75)
ax.legend()

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(source_healthcare))
ax.set_xticks(x)
ax.set_xticklabels(source_healthcare)

bar_width = 0.25
ax.bar(x - bar_width - bar_width, bad, width=bar_width, label='feel uncomfortable', color='#FFA500', edgecolor='black')
ax.bar(x - bar_width, good, width=bar_width, label='feel comfortable', color='#0000FF', edgecolor='black')
ax.bar(x, eh, width=bar_width, label='not care either way', color='#808080', edgecolor='black')
ax.bar(x + bar_width, no_answer, width=bar_width, label='No Answer', color='#FFFFFF', edgecolor='black')

ax.set_ylabel('Percentage (%)')
ax.set_title('Is it acceptable for AI to be used in Healthcare?')
plt.yticks(np.arange(0, 101, 5))
ax.set_ylim(0, 65)
ax.legend()

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
for i, size in enumerate(sizes_content_ai):
    ax.barh(source_labels_content_ai[i], size, color='blue', edgecolor='black')

ax.set_xlabel('Number in increments of 100 millions')
ax.set_title('Amount of AI-created images as of August 2023')
ax.set_xlim(0, 130)
ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
for i, size in enumerate(sizes_content_total):
    ax.barh(source_labels_content_total[i], size, color='orange', edgecolor='black')

ax.set_xlabel('Number in increments of 100 millions')
ax.set_title('Amount of content created as of August 2023')
ax.set_xlim(0, 16000)
ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
