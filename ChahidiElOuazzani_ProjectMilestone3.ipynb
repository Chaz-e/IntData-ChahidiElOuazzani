```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from colour import Color
```

```python
# graph 1 data
groups = ['GroupA', 'GroupB', 'GroupC', 'GroupD']
Group1 = np.array([10, 20, 10, 30])
Group2 = np.array([20, 25, 15, 25])
Group3 = np.array([12, 15, 19, 6])
Group4 = np.array([10, 29, 13, 19])

# graph 2 data
categories = ['A', 'B', 'C', 'D']
agree = [10, 15, 8, 12]
disagree = [-5, -2, -3, -8]
neutral = [5, 0, 2, -2]

use_cases = ['To write essays', 'To solve math\nproblems', 'To research\nnew topics']
not_acceptable = [57, 36, 13]
acceptable = [20, 39, 69]
not_sure = [23, 24, 18]

#more data
use_cases_2 = ['Dall-e 2', 'Midjourney', 'Adobe Firefly', 'Stable Diffusion']
AI_Images = [9.16, 9.64, 10, 125.9]

#Dear god how much data is there
red = Color("red")
colors = list(red.range_to(Color("green"),10))

data = [
 ['Number of AI-created images', 'DALL-E 2', 9.16],
 ['Number of AI-created images', 'Midjourney', 9.64],
 ['Number of AI-created images', 'Adobe Firefly', 10],
 ['Number of AI-created images', 'Models based on Stable Diffusion', 125.9],
 ['Amount of Content', 'Shutterstock Library', 3.86],
 ['Amount of Content', 'AI-Created Images', 150],
 ['Amount of Content', 'Images Uploaded to Instagram', 500],
 ['Amount of Content', 'Images Indexed on Google', 1360],
 ['Amount of Content', 'Pins Saved on Pinterest', 2400],
 ['Amount of Content', 'Images Uploaded to Facebook', 15000]]

source_labels = ['Number of AI-created images', 'Amount of Content']
sizes_A = [9.16, 9.64, 10, 125.9] 
sizes_B = [3.86, 150, 500, 1360, 2400, 15000]
```

```python
plt.bar(groups, Group1, color='r')
plt.bar(groups, Group2, bottom=Group1, color='y')
plt.bar(groups, Group3, bottom=Group1+Group2, color='b')
plt.bar(groups, Group4, bottom=Group1+Group2+Group3, color='g')

plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")

plt.show()
```

```python
fig, ax = plt.subplots()
y_indexes = np.arange(len(categories))

ax.barh(y_indexes, disagree, label='Disagree', color='red')
ax.barh(y_indexes, agree, label='Agree', color='green')

ax.spines['left'].set_position(('data', 0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_yticks(y_indexes)
ax.set_yticklabels(categories)

ax.set_xlabel('Values')
ax.set_title('Horizontal Bar Chart with Centered Axis')

ax.legend()

max_value = max(np.array(agree) + np.array(neutral) + np.array(disagree)) + 5
min_value = min(disagree) - 5
ax.set_xlim(min_value, max_value)

plt.show()
```

```python
fig, ax = plt.subplots()
y_indexes = np.arange(len(categories))

ax.barh(y_indexes, disagree, label='Disagree', color='red')
ax.barh(y_indexes, agree, label='Agree', color='green')
ax.barh(y_indexes, neutral, left=np.array(agree) + np.array(disagree), label='Neutral', color='gray')

ax.spines['left'].set_position(('data', 0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_yticks(y_indexes)
ax.set_yticklabels(categories)

ax.set_xlabel('Values')
ax.set_title('Horizontal Bar Chart with Centered Axis')

ax.legend()

max_value = max(np.array(agree) + np.array(neutral)) + 5
min_value = min(disagree) - 5
ax.set_xlim(min_value, max_value)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(use_cases))
ax.set_xticks(x)
ax.set_xticklabels(use_cases)

bar_width = 0.25
ax.bar(x - bar_width, not_acceptable, width=bar_width, label='Not acceptable', color='#8c564b')
ax.bar(x, acceptable, width=bar_width, label='Acceptable', color='#2ca02c')
ax.bar(x + bar_width, not_sure, width=bar_width, label='Not sure', color='#d62728')

ax.set_ylabel('Percentage (%)')
ax.set_title('U.S. Teens\' Perceptions of Acceptable ChatGPT Usage')
ax.legend()

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

```python
source_labels = ['Source A', 'Source B']
sizes = [0.8, 100]

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(source_labels, sizes, color=['blue', 'orange'])

for i, v in enumerate(sizes):
    ax.text(v + 2, i, f"{v:.1f}", color='black', va='center', fontsize=12)

ax.set_xlabel('Amount of Data')
ax.set_title('Data Created by Source A vs. Source B')
ax.set_xlim(0, 105)
ax.grid(axis='x', linestyle='--', alpha=0.7)
ax.axvline(x=1, color='gray', linestyle='--', linewidth=1)

plt.tight_layout()
plt.show()
```

```python
source_labels = ['Source A', 'Source B']
sizes_A = [0.2, 0.2, 0.2, 0.2]
sizes_B = [16.67, 16.67, 16.67, 16.67, 16.67, 16.67]

fig, ax = plt.subplots(figsize=(10, 6))
for i, size in enumerate(sizes_A):
    ax.barh(f'Source A Section {i + 1}', size, color='blue', edgecolor='black')

for i, size in enumerate(sizes_B):
    ax.barh(f'Source B Section {i + 1}', size, color='orange', edgecolor='black')

total_A = sum(sizes_A)
total_B = sum(sizes_B)

ax.text(total_A + 1, -0.5, f'Total: {total_A:.1f}', color='black', fontsize=12)
ax.text(total_B + 1, 4.5, f'Total: {total_B:.1f}', color='black', fontsize=12)
ax.set_xlabel('Amount of Data')
ax.set_title('Data Created by Source A vs. Source B')
ax.set_xlim(0, 105)
ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
